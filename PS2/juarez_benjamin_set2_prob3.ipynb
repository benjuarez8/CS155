```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
# Upload problem3data.txt to a 'data' folder in Google Colab, using the "Files"
# tab on the left.
# (You can also use urllib.request.urlretrieve, as in set 1)
train_file = 'https://raw.githubusercontent.com/charlesincharge/Caltech-CS155-2022/main/sets/set2/data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
# Lasso Regression
w = np.zeros(shape = (30,9))
alphas = np.linspace(0.01, 3, 30)

for i in range(np.size(w, 0)):
  ted_lasso = Lasso(alpha = alphas[i])
  ted_lasso.fit(X_train, y_train)
  w[i] = ted_lasso.coef_

plt.figure()
plt.title("Model weights (Lasso) vs. alpha")
plt.xlabel("Regularization strength (alpha)")
plt.ylabel("Model weights ($w_1,...,w_9$)")
plt.plot(alphas, w)
plt.legend(["$w_1$", "$w_2$", "$w_3$", "$w_4$", "$w_5$", "$w_6$", "$w_7$", 
             "$w_8$", "$w_9$"], loc=1)
plt.show()
```

```python
# Ridge Regression
w = np.zeros(shape = (10000,9))
alphas = [a for a in range(1, 10001)]

for i in range(np.size(w, 0)):
  oak_ridge = Ridge(alpha = alphas[i])
  oak_ridge.fit(X_train, y_train)
  w[i] = oak_ridge.coef_

plt.figure()
plt.title("Model weights (Ridge) vs. alpha")
plt.xlabel("Regularization strength (alpha)")
plt.ylabel("Model weights ($w_1,...,w_9$)")
plt.plot(alphas, w)
plt.legend(["$w_1$", "$w_2$", "$w_3$", "$w_4$", "$w_5$", "$w_6$", "$w_7$", 
             "$w_8$", "$w_9$"], loc=1)
plt.show()
```
